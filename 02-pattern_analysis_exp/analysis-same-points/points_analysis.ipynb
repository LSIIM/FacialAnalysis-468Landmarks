```python
!pip install matplotlib
import pandas as pd
import numpy as np
from tqdm import tqdm
from definitions import *
import os
import matplotlib.pyplot as plot
```

```python
users = os.listdir(PATTERN_TEST_PATH_SAME_POINT)
print(len(users))
```

```python
result_df = pd.DataFrame()
for i in range(2):
    for j in range(len(emotions_list)):
        tp = "0" + str(i)
        em = "0"+ str(j)
        result_df[em+'-'+tp] = np.zeros(468, dtype=float)
```

```python
result_df
```

```python

for user in tqdm(users):
    arqs = os.listdir(PATTERN_TEST_PATH_SAME_POINT+"/"+user)
    for arq in arqs:
        df = pd.read_csv(PATTERN_TEST_PATH_SAME_POINT+"/"+user+"/"+arq)
        em,tp = arq.split(".")[0].split("_")[0].split("-")
        result_df[em+"-"+tp] = np.array(result_df[em+"-"+tp] + df["distances"]/len(users))
        
```

```python
result_df.to_csv("./masks_average_dists.csv")
```

```python
for col in result_df:
    y=result_df[col]
    x=np.arange(468)

    plot.bar(x, y)
    plot.suptitle('Average values of distance to neutral face of ' + col)

    plot.savefig('./figs/'+col+'.png',dpi=400)
    plot.clf()
```

```python
for col in result_df:
    std = ((result_df[col] - result_df[col].mean())**2).mean()
    print(col,result_df[col].mean(),std)
```

```python
points_above_average = []
for col in result_df:
    points_list = []
    for i in range(468):
        if(result_df[col][i]>result_df[col].mean()):
            points_list.append(i)
    print(len(points_list))
    points_above_average.append(points_list)

```

```python
points_in_all_lists = []
for i in range(0,len(points_above_average)-1):
    if(points_in_all_lists == [] or points_in_all_lists == set()):
        points_in_all_lists = set(points_above_average[i]).intersection(points_above_average[i+1])
    else:
        points_in_all_lists =points_in_all_lists.intersection(points_above_average[i+1])
    print(len(points_in_all_lists))
```

```python
points_in_all_lists
```

```python
points_df = pd.DataFrame()
points_df["points_above_average"] = np.array(list(points_in_all_lists))
```
