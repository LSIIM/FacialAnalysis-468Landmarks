```python
import pandas as pd
import numpy as np
from tqdm import tqdm
from definitions import *
import os
import math
```

```python
users = os.listdir(PATTERN_TEST_PATH_SAME_POINT)
user_masks_means = []
points_df = pd.read_csv("./points_above_average.csv", index_col=0)
points_df
```

```python
for user in tqdm(users):
    mask_list = os.listdir(PATTERN_TEST_PATH_SAME_POINT+"/"+user)
    for mask in mask_list:
        mask_df = pd.read_csv(PATTERN_TEST_PATH_SAME_POINT+"/"+user+"/"+mask, index_col=0)
        mask_df = mask_df[mask_df["origins"].isin(points_df["points_above_average"].tolist())]
        achou = False
        for i,mask_means in enumerate(user_masks_means):
            if(mask_means[0]== mask.split("_")[0]):
                user_masks_means[i][1].append(mask_df["distances"].mean())
                achou = True
                break
        if(not achou):
            user_masks_means.append([mask.split("_")[0],[mask_df["distances"].mean()]])
    
        
```

```python
means_df = pd.DataFrame()
means_df["user_id"] = np.array(users)
for mean_col in user_masks_means:
    means_df[mean_col[0]] = np.array(mean_col[1])
```

```python
means_df = means_df.drop(columns=["00-00"])
```
